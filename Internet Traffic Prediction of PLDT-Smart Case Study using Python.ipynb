```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Classroom/MSDS 511 - Time Series Data Analysis and Forecasting 2023-2024/Internet Traffic Prediction Use Case Project/Dataset/Monthly Traffic and Usage.csv')
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# Set 'Date' as the index
data.set_index('Date', inplace=True)
```

```python
# Explore the first few rows and summary statistics
print(data.head())
print(data.describe())
```

```python
# Time series plot of total internet traffic
data['Total Internet Traffic (Tbps)'].plot(title='Total Internet Traffic Over Time')
plt.show()
```

```python
# Decomposition of time series to observe trends and seasonality
result = seasonal_decompose(data['Total Internet Traffic (Tbps)'], model='additive')
result.plot()
plt.show()
```
